```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('metrics.txt', sep=',', header=None)
```

```python
df = df.T
```

```python
df.columns = ["accuracy"]
```

```python
DF_HAR_1 = df.iloc[1:10]
DF_HAR_2 = df.iloc[10:19]
DF_FEM_1 = df.iloc[20:29]
DF_FEM_2 = df.iloc[29:38]
```

```python
print(len(DF_HAR_1), len(DF_HAR_2), len(DF_FEM_1), len(DF_FEM_2))
```

```python
fig, ax = plt.subplots(1, 2)
sns.barplot(x=DF_HAR_1.index, y=DF_HAR_1['accuracy'], ax=ax[0])
ax[0].set_title('HAR Vertical Accuracy')
ax[0].set_xlabel('%')

sns.barplot(x=DF_HAR_1.index, y=DF_HAR_2['accuracy'], ax=ax[1])
ax[1].set_title('HAR Horizontal Accuracy')
ax[1].set_xlabel('%')
```

```python
fig, ax = plt.subplots(1, 2)
sns.barplot(x=DF_HAR_1.index, y=DF_FEM_1['accuracy'], ax=ax[0])
ax[0].set_title('FEMNIST Vertical Accuracy')
ax[0].set_xlabel('%')

sns.barplot(x=DF_HAR_1.index, y=DF_FEM_2['accuracy'], ax=ax[1])
ax[1].set_title('FEMNIST Horizontal Accuracy')
ax[1].set_xlabel('%')
```

```python
DF_FEM_HOR = pd.read_csv('metrics_fem_hor.txt', sep=',', header=None).T
DF_FEM_VERT = pd.read_csv('metrics_fem_vert.txt', sep=',', header=None).T
```

```python
fig = sns.barplot(x=DF_FEM_HOR.index, y=DF_FEM_HOR[0])
fig.set_title('FEMNIST Horizontal Accuracy')
fig.set_xlabel('%')
```

```python
fig = sns.barplot(x=DF_FEM_VERT.index, y=DF_FEM_VERT[0])
fig.set_title('FEMNIST Vertical Accuracy')
fig.set_xlabel('%')
```

```python
fig, ax = plt.subplots(1, 2)
sns.barplot(x=DF_FEM_VERT.index, y=DF_FEM_VERT[0], ax=ax[0])
ax[0].set_title('FEMNIST Vertical Accuracy')
ax[0].set_xlabel('%')

sns.barplot(x=DF_FEM_HOR.index, y=DF_FEM_HOR[0], ax=ax[1])
ax[1].set_title('FEMNIST Horizontal Accuracy')
ax[1].set_xlabel('%')
```

```python
# FEMNIST Vertical Accuracy
df_fem_vert = pd.read_csv('metrics_fem_vert_10t.txt', sep=',', header=None)

df_grouped = df_fem_vert.T.groupby(np.arange(len(df.T)) // 10).agg(['mean', 'std'])

fig, ax = plt.subplots()
sns.barplot(x=df_grouped.index*10+10, y=df_grouped[0]['mean'], ax=ax)

ax.errorbar(df_grouped.index, df_grouped[0]['mean'], yerr=df_grouped[0]['std'], fmt='none', ecolor='black', capsize=5)
ax.set_title('FEMNIST Vertical Accuracy')
ax.set_xlabel('%')
ax.set_ylabel('Mean Accuracy Value')
plt.show()
```

```python
# FEMNIST Horizontal Accuracy
df_fem_hor = pd.read_csv('metrics_fem_hor_10t.txt', sep=',', header=None)

df_grouped = df_fem_hor.T.groupby(np.arange(len(df.T)) // 10).agg(['mean', 'std'])

fig, ax = plt.subplots()
sns.barplot(x=df_grouped.index*10+10, y=df_grouped[0]['mean'], ax=ax)

ax.errorbar(df_grouped.index, df_grouped[0]['mean'], yerr=df_grouped[0]['std'], fmt='none', ecolor='black', capsize=5)
ax.set_title('FEMNIST Vertical Accuracy')
ax.set_xlabel('%')
ax.set_ylabel('Mean Accuracy Value')
plt.show()
```
