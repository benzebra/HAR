```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('metrics.txt', sep=',', header=None)
```

```python
df = df.T
```

```python
df.columns = ["accuracy"]
```

```python
DF_HAR_1 = df.iloc[1:10]
DF_HAR_2 = df.iloc[10:19]
DF_FEM_1 = df.iloc[20:29]
DF_FEM_2 = df.iloc[29:38]
```

```python
print(len(DF_HAR_1), len(DF_HAR_2), len(DF_FEM_1), len(DF_FEM_2))
```

```python
fig, ax = plt.subplots(1, 2)
sns.barplot(x=DF_HAR_1.index, y=DF_HAR_1['accuracy'], ax=ax[0])
ax[0].set_title('HAR Vertical Accuracy')
ax[0].set_xlabel('%')

sns.barplot(x=DF_HAR_1.index, y=DF_HAR_2['accuracy'], ax=ax[1])
ax[1].set_title('HAR Horizontal Accuracy')
ax[1].set_xlabel('%')
```

```python
fig, ax = plt.subplots(1, 2)
sns.barplot(x=DF_HAR_1.index, y=DF_FEM_1['accuracy'], ax=ax[0])
ax[0].set_title('FEMNIST Vertical Accuracy')
ax[0].set_xlabel('%')

sns.barplot(x=DF_HAR_1.index, y=DF_FEM_2['accuracy'], ax=ax[1])
ax[1].set_title('FEMNIST Horizontal Accuracy')
ax[1].set_xlabel('%')
```
